```python
import csv 
import pandas as pd
import requests
from datetime import datetime
import math

def get_geolocations():
    
    now = datetime.now()
    print('Start Time: ', now)
    print('----------------------------------------------------------------------------------------')
    
    df = pd.read_csv('C:\\Users\\akshg\\OneDrive - infogroup\\Desktop\\OH_Input_File.csv',sep = '|')
    df
    df = df.astype('string')
    
    df['ClientGeolocation'] = df['TR_Long'] + ',' +df['TR_Lat']
    df['StoreGeolocation'] = df['TP_Long'] + ',' +df['TP_Lat']

    df = df.sort_values(by=['Unique_ID'], ascending=False)
    
    Unique_Id_df = df['Unique_ID'].copy().drop_duplicates()
    Unique_Id_df.columns = ['Unique_ID']
    
    
    for Unique_ID in Unique_Id_df:
        #print(Unique_ID)
        StoreGeolocation = ''
        ClientGeolocation = ''
        ClientGeolocation = df.loc[df['Unique_ID'] == Unique_ID, 'ClientGeolocation']
        StoreGeolocation = df.loc[df['Unique_ID'] == Unique_ID, 'StoreGeolocation']
        df10 = df.groupby('Unique_ID')
        s = StoreGeolocation.str.cat(sep=';')
        
        print('Unique_ID: ',Unique_ID)
        define_url(ClientGeolocation, s)
    print('----------------------------------------------------------------------------------------')
        

    print('Execution Time: ', datetime.now()-now)

def define_url(ClientGeolocation, StoreGeolocation):
    parameters = 'annotations=duration&skip_waypoints=true'
    source_index = 0
    mode_of_transport = 'driving'
    base_url = 'http://10.1.0.116:5000/table/v1/'
    
    call_url = base_url + mode_of_transport+'/'
    call_url = call_url + ClientGeolocation 
    
    call_url = call_url + ';' + StoreGeolocation + '?'
    call_url = call_url + 'sources='+ str(source_index) + '&'
    call_url = call_url + parameters
    call_url = call_url.drop_duplicates()
    
    for url in call_url:
        print(url)
        get_drivetime(url)

def get_drivetime(call_url):
    Drive_Time = []
    r = requests.get(call_url)
    response = r.json()
    for item in response['durations']:
        for i in item:
            if i==0:
                pass
            else:
                min =  (i/60.0)
                min = str(round(min, 2))
                Drive_Time.append(min)
                
        print(Drive_Time)
        print('----------------------------------------------------------------------------------------')
    
```
